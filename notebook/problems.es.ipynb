En este ejercicio vas a aplicar técnicas de estadística descriptiva sobre un conjunto de datos real: una base con 1000 películas populares del sitio IMDb.

A partir del título de cada película, vas a:
- Calcular la longitud del título (en caracteres).
- Obtener métricas estadísticas clave: media, mediana, moda, rango, varianza, desviación estándar, asimetría y curtosis.
- Identificar los títulos más largos y más cortos.
- Visualizar la distribución de estas longitudes con un histograma y una curva de densidad.


```python
# Importar librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis, mode

# Cargar dataset de IMDb
url = "../assets/imdb_1000.csv"
df = pd.read_csv(url)
df
```

```python
# Importar librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis, mode

# Cargar dataset de IMDb
url = "../assets/imdb_1000.csv"
df = pd.read_csv(url)

# Crea una nueva columna llamada 'title_length' con la longitud (número de caracteres) de cada título
df['title_length'] = df['title'].str.len()
df
```

```python
# Calcula media, mediana y moda de la columna 'title_length'
# Mean
mean = df['title_length'].mean()

# Median
median = df['title_length'].median()

# Mode
mode = df['title_length'].mode()

f'La media es {mean:.2f}%, la mediana es {median} y la moda es {mode.item()}.'
```

```python
# Calcula el rango, la varianza y la desviación estándar de 'title_length'
var_title = df['title_length'].var()
std_title = df['title_length'].std()
range_titles = df['title_length'].max() - df['title_length'].min()

f'La varianza es {var_title:.2f} y la desviación estándar es {std_title:.2f}. El rango es {range_titles}.'
```

```python
# Calcula asimetría y curtosis de la columna 'title_length'
skewness_title = df['title_length'].skew()
kurtosis_title = df['title_length'].kurt()

f'La asimetría es {skewness_title:.2f} y la curtosis es {kurtosis_title:.2f}.'
```

```python
# Imprime el título más corto y el título más largo según su longitud
length_titles = df['title'].str.len()

f'El título más largo es {length_titles.max()} y el títlo más corto es {length_titles.min()}'
```

```python
# Histogramas de curvas de densidades:
plt.figure(figsize=(10, 6))
plt.hist(df['title_length'], bins=20, density=True, alpha=0.6, color='g')       
df['title_length'].plot(kind='kde', color='blue')
plt.title('Distribución de la longitud de los títulos')
plt.xlabel('Longitud del título')
plt.ylabel('Densidad')
plt.show()
```

```python
# Curvas de densidades:
plt.figure(figsize=(10, 6))
sns.histplot(df['title_length'], bins=20, kde=True, color='b')
plt.title('Distribución de la longitud de los títulos')
plt.xlabel('Longitud del título')
plt.ylabel('Densidad')
plt.show()
```

## Ejercicio 2 – Cálculo manual de desviación estándar

En este ejercicio vas a calcular manualmente la desviación estándar, seleccionando cinco títulos reales del dataset de películas que cargaste antes.


```python
# Paso 1: Crea una lista con 5 valores reales de df['title_length'], por ejemplo: [10, 13, 14, 18, 22]
list_title_lengths = df['title_length'].head(5).tolist()
list_title_lengths
```

```python
# Calcula la media de los 5 valores usando pandas o  numpy
mean_title_lengths = np.mean(list_title_lengths)
f'La media es de {mean_title_lengths.round()} caracteres'
```

```python
# Calcula las diferencias al cuadrado con respecto a la media:
squared_differences = [(x - mean_title_lengths) ** 2 for x in list_title_lengths]
screen_results = pd.Series(squared_differences)
f'La diferencia al cuadrado por fila con respecto a la media es {screen_results.iloc[0:4].round(2).to_list()} caracteres' 
```

```python
# Calcula la varianza (promedio de las diferencias al cuadrado) y su raíz cuadrada para obtener la desviación
variance = np.mean(squared_differences)
std_deviation = np.sqrt(variance)
f'La varianza es {variance.round(2)} y la desviación estándar es {std_deviation.round(2)} caracteres'
```
